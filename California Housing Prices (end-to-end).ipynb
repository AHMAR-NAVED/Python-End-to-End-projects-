```python
from sklearn.preprocessing import StandardScaler
```

```python
std_sclr = StandardScaler()

scld_data = std_sclr.fit_transform(new_data)
```

```python
scld_data
```

```python
scld_data = pd.DataFrame(scld_data, columns=new_data.columns)
```

```python
scld_data
```

```python
scld_data.info()
```

```python
scld_data.isnull().sum()
```

```python
x = scld_data.drop(['median_house_value'], axis = 1)
y = scld_data['median_house_value']
```

```python
y.head(10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20)
```

```python
from sklearn.linear_model import LinearRegression
lin_model = LinearRegression()
lin_model.fit(x_train,y_train)
lin_model.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg_model = RandomForestRegressor()
reg_model.fit(x_train,y_train)
reg_model.fit(x_train,y_train)
```

```python
from sklearn.neighbors import KNeighborsRegressor
reg_model = KNeighborsRegressor()
reg_model.fit(x_train,y_train)
reg_model.score(x_test,y_test)
```
